```python
df = spark.table("stock_data")

display(df.limit(5))
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

window_5 = Window.orderBy("Date").rowsBetween(-4,0)

df = df.withColumn(
    "SMA_5",
    F.avg("Close").over(window_5)
)
```

```python
# simplified example
df = df.withColumn(
    "daily_return",
    (F.col("Close") - F.col("Open")) / F.col("Open")
)
```

```python
df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("stock_features")
```
