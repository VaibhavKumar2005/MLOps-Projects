# Databricks Serving Demo (Optimized Setup)

This notebook starts with Databricks runtime and Spark tuning defaults for faster startup, stable execution, and reproducible model loading.

```python
from pyspark.sql import SparkSession
import mlflow

spark = SparkSession.getActiveSession() or SparkSession.builder.getOrCreate()

# Runtime-aware defaults for Databricks execution efficiency.
spark.conf.set("spark.sql.adaptive.enabled", "true")
spark.conf.set("spark.sql.execution.arrow.pyspark.enabled", "true")

print("Databricks setup complete.")
print(f"Spark version: {spark.version}")
```

```python
import mlflow.pyfunc

model = mlflow.pyfunc.load_model(
    "models:/stock_model/Production"
)
```

```python
import pandas as pd

sample = pd.DataFrame([{"Close": 150.0, "Returns": 0.01}])

prediction = model.predict(
    sample
)

print(prediction)
```
